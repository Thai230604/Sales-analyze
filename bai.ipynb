```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='most_frequent')
l = ['ADDRESSLINE1', 'ADDRESSLINE2', 'STATE', 'POSTALCODE', 'TERRITORY']
df[l] = imputer.fit_transform(df[l])
```

```python
df.isnull().sum()
```

```python
for column in df.columns:
    if df[column].dtype != 'object':
        sns.boxplot(df[column])
        plt.show()

        q1 = df[column].quantile(0.25)
        q3 = df[column].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        df= df[(df[column] < upper_bound) & (df[column] > lower_bound)]
        sns.boxplot(df[column])
        plt.show()
```

```python
features = ['QUANTITYORDERED', 'PRICEEACH', 'ORDERLINENUMBER', 'PRODUCTLINE', 'MSRP', 'DEALSIZE', 'TERRITORY', 'MONTH_ID', 'QTR_ID', 'YEAR_ID']
df[features].head()
```

```python
df['DEALSIZE'].unique()
```

```python
num_transform = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])
sizes = ['Small', 'Medium', 'Large']
ord_transform = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OrdinalEncoder(categories=[sizes]))
])
nom_transform = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder())
])
```

```python
preprocessor = ColumnTransformer([
    ('num', num_transform, ['QUANTITYORDERED', 'PRICEEACH', 'ORDERLINENUMBER', 'MSRP', 'MONTH_ID', 'QTR_ID', 'YEAR_ID']),
    ('ord', ord_transform, ['DEALSIZE']),
    ('nom', nom_transform, ['PRODUCTLINE', 'TERRITORY'])
])
```

```python
df[features].head()
```

```python
new_df = df[features]
new_df.head()
```

```python
sizes = ['Small', 'Medium', 'Large']
new_df['DEALSIZE'] = OrdinalEncoder(categories=[sizes]).fit_transform(new_df[['DEALSIZE']])
```

```python
new_df
```

```python
new_df = pd.get_dummies(new_df, columns=['PRODUCTLINE', 'TERRITORY'], drop_first=True)
```

```python
x = new_df
y = df['SALES']
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
reg = LazyRegressor(verbose=0, ignore_warnings=False, custom_metric=None)
models, predictions = reg.fit(x_train, x_test, y_train, y_test)
models
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor
lgbm = GradientBoostingRegressor()
lgbm.fit(x_train, y_train)
y_pred = lgbm.predict(x_test)
from sklearn.metrics import mean_squared_error, r2_score
mean_squared_error(y_test, y_pred)
r2_score(y_test, y_pred)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)
y_pred = lin_reg.predict(x_test)
r2_score(y_test, y_pred)
```

```python
from sklearn.svm import SVR
svc = SVR()
svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)
r2_score(y_test, y_pred)
```
